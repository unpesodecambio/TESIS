## Operaciones básicas con numpy

```python
import numpy as np

# Crear arrays unidimensionales
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 6, 7, 8])

# Crear arrays bidimensionales (matrices)
arr3 = np.array([[1, 2], [3, 4]])
arr4 = np.array([[5, 6], [7, 8]])
```

```python
def cargar_archivo(ruta_archivo):
    if ruta_archivo.endswith('.csv'):
        datos = pd.read_csv(ruta_archivo, dtype= {'ENTIDAD': str,'MUN':str,'LOC':str,'AGEB':str,'MZA':str})
        print("se ha cargado el CSV")
    elif ruta_archivo.endswith('.xlsx'):
        datos = pd.read_excel(ruta_archivo)
        print("se ha cargado el XLSX")
    else:
        raise ValueError("Error al cargar el archivo. Verifica la ruta y la extensión del mismo.")
    return datos

censo_2020 = cargar_archivo(ruta_censo_csv)

# Mostrar las primeras filas
print("Datos cargados:")
censo_2020.head()
```
